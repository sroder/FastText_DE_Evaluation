```python
file = 'Datenbersicht - Artificial Intelligence Sales Guru (m_w_d) .txt'
```

```python
with open(file, 'r') as text:
    text = text.read()
```

```python
def ExpText2Token(raw_text):
    raw_text = raw_text.replace('\\','')
    raw_text = raw_text.replace('/','')
    raw_text = raw_text.replace('c("','')
    raw_text = raw_text.replace('\r\n','')
    raw_text = raw_text.replace('NA','')
    raw_text = raw_text.replace(')','')
    raw_text = raw_text.replace('(','')
    raw_text = raw_text.replace('"','')
    raw_text = raw_text.replace(',','')
    raw_text = raw_text.replace('\r',' ')
    raw_text = raw_text.replace('-',' ')
    raw_text = raw_text.replace(':','')
    tokens = []
    for i in raw_text.split(' '):
        tokens.append(i.strip())
    return list(set(tokens))
```

```python
req_tokens = []
for i in text.split('\n'):
    temp = ExpText2Token(i)
    req_tokens = req_tokens + temp
req_tokens =list(set(req_tokens))
```

```python
import pandas as pd
import csv

words = pd.read_table(r'C:\Users\kumarsa\Desktop\Experiment\Multilingual embeddings\multilingual_embeddings.de', sep=" ", index_col=0, header=None, quoting=csv.QUOTE_NONE)
```

```python
def vec(w):
        return words.loc[w].as_matrix()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
first = vec('java').reshape(1, -1)
second = vec('c').reshape(1, -1)
print(cosine_similarity(first, second)[0])
```
